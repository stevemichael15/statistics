- Measures of Central Tendency:
->  Mean: The average of a data set. It's calculated by adding all numbers and dividing by the count of numbers.

Use When: Data is symmetrically distributed without outliers.

->  Median: The middle value when data is ordered from least to greatest.
Example: The median of 2, 4, 6 is 4. For 2, 4, 6, 8, the median is (4+6)/2 = 5.
Use When: Data has outliers or is skewed.

->  Mode: The most frequently occurring value(s) in a data set.
Example: The mode of 2, 4, 4, 6 is 4.
Use When: Identifying the most common value is necessary.

4.   What is a box plot, and what can it tell you about the distribution of data?

- A box plot, also known as a box-and-whisker plot, is a graphical representation that shows the distribution of a data set. It provides a summary of the data's central tendency, dispersion, and shape.

        Components of a Box Plot:

->  Box: Represents the interquartile range (IQR), which contains the middle 50% of the data.

->  Median Line: Inside the box, the line represents the median (the middle value of the data set).

->  Whiskers: Extend from the box to the smallest and largest values within 1.5 times the IQR from the lower and upper quartiles.

->  Outliers: Points outside the whiskers are considered outliers and are often plotted as individual points.

What It Tells You:
Median: Indicates the central value of the data set.
->  IQR: Shows the spread of the middle 50% of the data.
->  Whiskers: Indicate the range of the data, excluding outliers.
->  Outliers: Points beyond the whiskers, indicating unusually high or low values.
->  Symmetry: The box plot can show if the data is symmetric or skewed.
    Symmetric Data: The median line is near the center of the box, and whiskers are roughly equal in length.
    Skewed Data: The median line is closer to one end of the box, and whiskers are unequal in length.

5.  Discuss the role of random sampling in making inferences about populations

- Random Sampling: A technique where each member of a population has an equal chance of being selected to be part of the sample.

        Role in Making Inferences:
->  Representation: Ensures the sample represents the population, capturing its characteristics accurately. This minimizes bias and allows for generalizations to be made about the entire population.

->  Accuracy: Improves the accuracy of estimates and predictions. Random samples reduce the risk of systematic errors, providing a clearer picture of the population's attributes.

->  Variability Measurement: Helps in measuring the variability within the population. By analyzing the sample, statisticians can infer the range and distribution of data in the population.


6.   Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

- Skewness refers to the asymmetry or lack of symmetry in the distribution of data. It indicates whether the data points are distributed more to one side of the mean than the other.

                Types of Skewness
i.  Positive Skew (Right Skew):
Characteristics: The tail on the right side of the distribution is longer or fatter. Most data points are clustered on the left.
Example: Income distribution, where a few people earn significantly more than the majority.
Interpretation: The mean is greater than the median.

ii. Negative Skew (Left Skew):
Characteristics: The tail on the left side of the distribution is longer or fatter. Most data points are clustered on the right.
Example: Age at retirement, where most people retire at a standard age, but some retire much earlier.
Interpretation: The mean is less than the median.

iii.  Zero Skew (Symmetric):
Characteristics: The distribution is symmetric, and the tails on both sides are equally balanced.
Example: Heights of adult men in a population.
Interpretation: The mean and median are equal or very close.

Effect of Skewness on Data Interpretation
->  Central Tendency:
    In a positively skewed distribution, the mean is higher than the median.
    In a negatively skewed distribution, the mean is lower than the median.
    The median is often preferred over the mean in skewed distributions because it is less affected by outliers.

->  Spread and Dispersion:
    Skewness affects the range and standard deviation.
    Positively skewed data may have higher variability, while negatively skewed data may show lower variability.

->  Outliers:
    Skewness indicates the presence of outliers.
    Positive skewness suggests potential high-value outliers.
    Negative skewness suggests potential low-value outliers.

7.   What is the interquartile range (IQR), and how is it used to detect outliers?

- Interquartile Range (IQR)
The Interquartile Range (IQR) is a measure of statistical dispersion, representing the middle 50% of a data set. It’s calculated as the difference between the first quartile (Q1) and the third quartile (Q3)

Steps to Calculate IQR:
>Order the Data: Arrange the data points in ascending order.
>Find Q1: The first quartile (25th percentile).
>Find Q3: The third quartile (75th percentile).
>Calculate IQR: Subtract Q1 from Q3.

->  Detecting Outliers with IQR
Outliers are data points that lie significantly outside the typical range of the data. The IQR is used to detect outliers by identifying data points that fall below Q1 - 1.5IQR or above Q3 + 1.5IQR.

Steps:
>Calculate IQR.
>Determine the lower bound: Q1 - 1.5*IQR.
>Determine the upper bound: Q3 + 1.5*IQR.
>Identify outliers: Data points outside these bounds are considered outliers.


8.   Discuss the conditions under which the binomial distribution is used

- Binomial Distribution
The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials of a binary (two-outcome) experiment. Here are the conditions under which the binomial distribution is used:

Fixed Number of Trials (n):
>The experiment or process is repeated a specific number of times.
>Example: Flipping a coin 10 times.

Binary Outcomes:
>Each trial has exactly two possible outcomes: success or failure.
>Example: Each coin flip results in either heads (success) or tails (failure).

Independence:
>The outcome of each trial is independent of the others.
>Example: The result of each coin flip does not affect the result of the next flip.

Constant Probability (p):
>The probability of success (p) is the same for each trial.
>Example: The probability of flipping heads remains 0.5 for each coin flip.

9.   Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

- Normal Distribution
The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric around its mean. It is characterized by its bell-shaped curve.

Properties of Normal Distribution:
>Symmetry: The distribution is symmetric about the mean. This means that the left and right sides of the curve are mirror images.
>Mean, Median, and Mode: In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.
>Bell-Shaped Curve: The distribution forms a bell-shaped curve, where most data points are concentrated around the mean.
>Asymptotic: The tails of the distribution approach, but never touch, the horizontal axis.
>Total Area: The total area under the curve is equal to 1, representing the total probability of all outcomes.

- Empirical Rule (68-95-99.7 Rule)
The empirical rule describes how data is distributed in a normal distribution. It states that:
>68%: Approximately 68% of the data falls within one standard deviation (σ) of the mean (μ).
>95%: Approximately 95% of the data falls within two standard deviations (2σ) of the mean.
>99.7%: Approximately 99.7% of the data falls within three standard deviations (3σ) of the mean.

11.  Explain what a random variable is and differentiate between discrete and continuous random variables.

- A random variable is a numerical value that represents the outcomes of a random experiment or process. It assigns a numerical value to each possible outcome of the random process.

Types of Random Variables
->  Discrete Random Variable
Definition: A discrete random variable takes on a countable number of distinct values. These values are often whole numbers.

Examples:
>Number of heads in 10 coin flips (e.g., 0, 1, 2,..., 10)
>Number of students present in a class on a given day (e.g., 0, 1, 2,...)

Characteristics:
>Values are countable and finite.
>Probability distribution can be represented with a probability mass function.

->  Continuous Random Variable
Definition: A continuous random variable takes on an infinite number of possible values within a given range. These values are often measured, rather than counted.

Examples:
>Height of students in a class (e.g., any value between 150 cm and 200 cm)
>Time taken to run a marathon (e.g., any value from 2 hours to 5 hours)

Characteristics:
>Values are uncountable and can take any value within an interval.
>Probability distribution is represented with a probability density function.
